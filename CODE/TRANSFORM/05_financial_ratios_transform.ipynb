## Computing secondary metrics - financial ratios

```python
def get_ratios(row):
    # Ukazovatele likvidity
    row['L1'] = ( row['CASH'] ) / ( row['CL'] ) if row['CL'] != 0 else None
    row['L2'] = ( row['CASH'] + row['CC'] ) / ( row['CL'] ) if row['CL'] != 0 else None
    row['L3'] = ( row['CASH'] + row['CC'] + row['INV'] ) / ( row['CL'] ) if row['CL'] != 0 else None
    row['NCL_CF'] = ( row['NCL'] ) / ( row['CF_NETTO'] ) if row['CF_NETTO'] != 0 else None
    row['CASH_TA'] = ( row['CASH'] ) / ( row['TA'] ) if row['TA'] != 0 else None

    # Ukazovatele aktivity
    row['INV_COGS'] = ( row['INV'] ) / ( row['costs_of_merchandise_sold'] ) if row['costs_of_merchandise_sold'] != 0 else None
    row['CC_SAL'] = ( row['CC'] ) / ( row['SAL'] ) if row['SAL'] != 0 else None
    row['TL_COST'] = row['TL'] / ( row['COST'] ) if row['COST'] > 0 else None
    row['CL_COST'] = row['CL'] / ( row['COST'] ) if row['COST'] > 0 else None
    row['SAL_TA'] = ( row['SAL'] ) / ( row['TA'] ) if row['TA'] != 0 else None

    # Ukazovatele zadlženosti
    row['EQ_TA'] = row['EQ'] / ( row['TA'] ) if row['TA'] > 0 else None
    row['EQ_TL'] = ( row['EQ'] ) / ( row['EQ'] + row['TL'] ) if ( row['EQ'] + row['TL'] ) != 0 else None
    row['CL_CC'] = ( row['CL'] ) / ( row['CC'] ) if row['CC'] != 0 else None    
    row['LTC_NCA'] = ( row['EQ'] + row['TL'] - row['CL'] ) / ( row['TA'] - row['CA'] )  if ( row['TA'] - row['CA'] ) != 0 else None
    row['TZ'] = ( row['TL'] - row['CASH'] - row['CC'] - row['INV'] ) / row['CF_NETTO']  if row['CF_NETTO'] != 0 else None
    row['CF_TL'] = ( row['CF_NETTO'] ) / ( row['TL'] ) if row['TL'] != 0 else None
    row['CL_TA'] = ( row['CL'] ) / ( row['TA'] ) if row['TA'] != 0 else None
    row['NCL_TA'] = ( row['NCL'] ) / ( row['TA'] ) if row['TA'] != 0 else None

    # Ukazovatele rentability
    row['EBT_REV'] = ( row['EBIT'] ) / ( row['REV'] ) if row['REV'] != 0 else None
    row['ROS'] = ( row['EAT'] ) / ( row['SAL'] ) if row['SAL'] != 0 else None
    row['EAT_REV'] = ( row['EAT'] ) / ( row['REV'] ) if row['REV'] != 0 else None
    row['ROA'] = ( row['EAT'] ) / ( row['TA'] ) if row['TA'] != 0 else None
    row['ROE'] = ( row['EAT'] ) / ( row['EQ'] ) if row['EQ'] != 0 else None
    row['CF_TA'] = ( row['CF_NETTO'] ) / ( row['TA'] ) if row['TA'] != 0 else None
    row['CF_SAL'] = ( row['CF_NETTO'] ) / ( row['SAL'] ) if row['SAL'] != 0 else None

    # Ukazovatele efektivity
    row['LABOR_EAT'] = row['personnel_costs'] / ( row['EAT'] ) if row['EAT'] != 0 else None    
    row['LABOR_REV'] = row['personnel_costs'] / ( row['REV'] ) if row['REV'] != 0 else None
   

    return row


records = []
k = 0
size = df.shape[0]

for row_dict in df.to_dict(orient="records"):
    # if k >= 100:
    #     break

    if (k + 1) % 50000 == 0:
        print(f'{k + 1} / {size}')

    k += 1

    records.append(get_ratios(row_dict))


df_extended = pd.DataFrame(records)

ratios = ['L1',
       'L2', 'L3', 'NCL_CF', 'CASH_TA', 'INV_COGS', 'CC_SAL', 'TL_COST',
       'CL_COST', 'SAL_TA', 'EQ_TA', 'EQ_TL', 'CL_CC', 'LTC_NCA', 'TZ',
       'CF_TL', 'CL_TA', 'NCL_TA', 'EBT_REV', 'ROS', 'EAT_REV', 'ROA', 'ROE',
       'CF_TA', 'CF_SAL', 'LABOR_EAT', 'LABOR_REV']



df = None
```
